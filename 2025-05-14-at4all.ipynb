# gathering data from the AT4All database

it is a graphql endpoint

````python
    import midgy, copy, pandas, requests
    from toolz.curried import *
````

````python
%%
the post `json` data to request.

    params =\
```json
{"operationName":"ITEM_SEARCH_GROUPED","variables":{"programId":null,"searchPhrase":"","federalCategoryId":0,"skip":0,"take":15,"showAdminCheckouts":null,"showRestrictedItems":null,"sortOptions":"NameAsc","transactionTypeID":null,"fromZip":0,"hideExpired":true},"query":"query ITEM_SEARCH_GROUPED($categoryIDs: [Int], $federalCategory: String, $programId: Int, $searchPhrase: String, $federalCategoryId: Int, $skip: Int, $take: Int, $showAdminCheckouts: Boolean, $showRestrictedItems: Boolean, $sortOptions: SortType, $fromZip: Int, $transactionTypeID: Int, $hideExpired: Boolean) {\n  allItemsGrouped(\n    categoryIDs: $categoryIDs\n    federalCategory: $federalCategory\n    programId: $programId\n    searchPhrase: $searchPhrase\n    federalCategoryId: $federalCategoryId\n    skip: $skip\n    take: $take\n    showAdminCheckouts: $showAdminCheckouts\n    showRestrictedItems: $showRestrictedItems\n    sortOptions: $sortOptions\n    fromZip: $fromZip\n    transactionTypeID: $transactionTypeID\n    hideExpired: $hideExpired\n  ) {\n    ItemGroups {\n      ItemCount\n      ItemIDs\n      ItemName\n      Description\n           FederalReportingCategoryName\n      ProgramID\n      ProgramName\n      TransactionTypeName\n      TransactionTypeID\n      CityDisplay\n      Make\n      Model\n      AskingPrice\n      Availability\n      HasAdminCheckout\n      InventoryTag\n      __typename\n    }\n    TotalCount\n    RowCount\n    __typename\n  }\n}"}
```
````

````python
    p = copy.deepcopy(params)

    p["variables"]["take"] = 10
    p["variables"]["skip"] = 10

    response = requests.post(
        "https://www.at4all.com/api/data",
        json = p
    )
    response
````

````python
    pipe(
        response.json(),
        get("data"), 
        get("allItemsGrouped"), 
        get("ItemGroups"), 
        pandas.Series
    ).apply(pandas.Series).T
````
